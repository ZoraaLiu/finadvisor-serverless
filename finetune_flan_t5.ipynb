### Deployment Steps (manual in AWS Console)
1. Go to AWS Lambda → Create Function.
2. Runtime: Python 3.10, Memory: 1024MB+, Timeout: 30s.
3. Upload `lambda_package.zip`.
4. Set handler to `lambda_handler.lambda_handler`.
5. Connect via API Gateway → POST `/generate`.
